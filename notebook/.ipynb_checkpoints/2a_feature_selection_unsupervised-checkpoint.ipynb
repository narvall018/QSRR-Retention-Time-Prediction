# Unsupervised feature selection

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import os

# Créer le dossier pour sauvegarder les résultats dans 'figures/pca_results'
os.makedirs('figures/pca_results', exist_ok=True)

# Fonction pour PCA et visualisation
def pca_analysis_and_plot(df, title, target_col='RT', id_col='Molecule_Name'):
    # Préparation des données
    y = df[target_col]
    X = df.drop([target_col, id_col], axis=1)
    
    # Remplir les valeurs manquantes avec la moyenne de chaque colonne
    X = X.fillna(X.mean())
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # PCA
    pca = PCA(n_components=0.95)
    pca_result = pca.fit_transform(X_scaled)
    n_components = pca_result.shape[1]
    
    # Plot
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    
    # Variance cumulée
    cumsum = np.cumsum(pca.explained_variance_ratio_)
    ax[0].plot(range(1, len(cumsum) + 1), cumsum, 'bo-')
    ax[0].axhline(y=0.95, color='r', linestyle='--', label='95% threshold')
    ax[0].set_xlabel('Number of Components')
    ax[0].set_ylabel('Cumulative Explained Variance')
    ax[0].set_title(f'{title}\n{n_components} components for 95% variance')
    ax[0].legend()
    
    # Top 20 descriptors loadings
    loadings = pd.DataFrame(
        pca.components_[0],
        columns=['PC1'],
        index=X.columns
    )
    loadings['Abs'] = abs(loadings['PC1'])
    top_loadings = loadings.nlargest(20, 'Abs')
    
    colors = ['red' if x < 0 else 'blue' for x in top_loadings['PC1']]
    ax[1].barh(range(len(top_loadings)), top_loadings['PC1'], color=colors)
    ax[1].set_yticks(range(len(top_loadings)))
    ax[1].set_yticklabels(top_loadings.index, fontsize=8)
    ax[1].set_xlabel('PC1 Loading')
    ax[1].set_title('Top 20 Descriptors in PC1')
    
    plt.tight_layout()
    
    # Normaliser le titre pour un nom de fichier valide
    sanitized_title = title.replace(" ", "_").replace("/", "_")
    # Sauvegarder chaque figure dans 'figures/pca_results' avec le nom du dataset
    plt.savefig(f'figures/pca_results/{sanitized_title}_pca_plot.png', dpi=300, bbox_inches='tight')
    plt.show()  # Afficher le graphique
    
    return n_components, pca.explained_variance_ratio_[0]

# Analyse pour chaque type de descripteur
datasets = {
   'RDKit 2D': descriptors_2d,
   'ECFP': descriptors_ecfp,
   'MACCS': descriptors_maccs,
   'MQN': descriptors_mqn,
   'Mordred 2D/3D': descriptors_23d,
   'Combined': all_df
}

results = {}
for name, df in datasets.items():
    n_comp, var_ratio = pca_analysis_and_plot(df, name)
    results[name] = {'n_components': n_comp, 'first_pc_variance': var_ratio}

# Convertir les résultats en DataFrame et afficher
summary_df = pd.DataFrame(results).T
print("\nSummary of PCA results:")
print(summary_df)

# Sauvegarder le résumé dans le dossier 'figures/pca_results'
summary_df.to_csv('figures/pca_results/summary_pca_results.csv')
```

```python
import os
import pandas as pd

# Créer un dossier pour les données PCA si il n'existe pas
output_dir = "data/feature_selection/unsupervised"
os.makedirs(output_dir, exist_ok=True)

# Fonction pour sauvegarder les données PCA
def save_pca_data(pca_result, y, n_comp, pca, features, name):
    # Nettoyer le nom pour les fichiers
    sanitized_name = name.replace(" ", "_").replace("/", "_")
    
    # Créer DataFrame avec les composantes principales
    pca_df = pd.DataFrame(
        pca_result[:, :n_comp],  # Prendre seulement les composantes nécessaires pour 95%
        columns=[f'PC{i+1}' for i in range(n_comp)]
    )
    
    # Ajouter la target
    pca_df['RT'] = y.reset_index(drop=True)
    
    # Sauvegarder les données transformées
    pca_df.to_parquet(os.path.join(output_dir, f'{sanitized_name}_pca.parquet'))
    
    # Sauvegarder les loadings
    loadings = pd.DataFrame(
        pca.components_[:n_comp].T,
        columns=[f'PC{i+1}' for i in range(n_comp)],
        index=features
    )
    loadings.to_csv(os.path.join(output_dir, f'{sanitized_name}_loadings.csv'))
    
    print(f"Sauvegardé {name}:")
    print(f"- Données PCA shape: {pca_df.shape}")
    print(f"- Loadings shape: {loadings.shape}\n")

# Appliquer la fonction pour chaque jeu de données
for name, df in datasets.items():
    y = df['RT']
    X = df.drop(['RT', 'Molecule_Name'], axis=1)
    
    # Préparation des données et PCA
    X = X.fillna(X.mean())  # Remplissage des NaN
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    pca = PCA(n_components=0.95)
    pca_result = pca.fit_transform(X_scaled)
    n_comp = pca_result.shape[1]
    
    # Sauvegarder les données PCA et les loadings
    save_pca_data(pca_result, y, n_comp, pca, X.columns, name)
```
